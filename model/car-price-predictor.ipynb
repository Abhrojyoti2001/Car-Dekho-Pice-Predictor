```python
mean_mileage = df[df['mileage'] !=0]['mileage'].median()
mean_mileage
```

```python
df['mileage'] = np.where(df['mileage'] == 0,mean_mileage,df['mileage'])
```

```python
# spliting x and y for train and test
X = df.drop(columns=['selling_price'])
y = df['selling_price']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=2)
```

```python
# Ordinal encoding on Owner column and Nominal encoding on fuel, seller_type, transmission, brand columns
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer

tnf = ColumnTransformer(
    [('ordinal',OrdinalEncoder(),['owner']),('nominal',OneHotEncoder(drop='first',sparse=False),['fuel','seller_type','transmission','brand'])],
    remainder='passthrough')
```

```python
# PolynomialFeatures
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)

# LinearRegression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# import r2_score
from sklearn.metrics import r2_score
```

```python
# Pipeline
from sklearn.pipeline import Pipeline
pipe = Pipeline([('col-transformer',tnf),('poly',poly),('lr',lr)])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```
